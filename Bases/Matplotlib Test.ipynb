```python
"""Use NumPy to generate random data"""
import numpy as np
x=np.linspace(0,10,25)
y=x*x+2
print(x)
print(y)
print(np.array([x,y]))
print(np.array([x,y]).reshape(25,2))
```

```python
fig=plt.figure()
axis=fig.add_axes([0.5, 0.1, 0.8,0.8]) # Control de left, right, width, height of the canvas (from 0 to 1)
axis.plot(x,y,'r')
```

```python
# again we can draw subgraphs
fig, axes=plt.subplots(nrows=2,ncols=1) #Submap is of 1 raw, 2 columns
for ax in axes:
    ax.plot(x,y,'r')
```

```python
# We can draw a picture, or graph, inside another graph 
fig=plt.figure()
# Control de left, right, width, height of the canvas (from 0 to 1)
axes1=fig.add_axes([0.1,0.1,0.8,0.8]) #big axes
axes2=fig.add_axes([0.2,0.5,0.4,0.3]) #small canvas
axes1.plot(x,y,'r')
axes2.plot(y,x,'b')
```

```python
fig,ax=plt.subplots(dpi=100) 
# Changing of size
ax.plot(x,x+4,color='blue',linewidth=0.1)
ax.plot(x,x+6,lw=1,linestyle='-.')
ax.plot(x,x+8,linewidth=1.5,marker='o',markerfacecolor='blue')
ax.plot(x,x+10,linewidth=2,marker='s',markersize=10)
ax.plot(x,x+10,linewidth=2)

line, =ax.plot(x,x+12,color='black',lw=1.50)
line.set_dashes([5,12,15,10]) # Size space size space size space
line, =ax.plot(x,x+14,color='black',lw=1.50)
line.set_dashes([5,14,15,1])
line, =ax.plot(x,x+16,color='black',lw=1.50)
line.set_dashes([5,16,5,10])
```

```python
# Set the canvas grid and axis range
fig, Axes=plt.subplots(1,2,figsize=(10,5))

Axes[0].plot(x,x**2,x,x**3,lw=2)
Axes[0].grid(True)

Axes[1].plot(x,x**2,x,x**3,lw=2)
Axes[1].set_ylim([0,60])
Axes[1].set_xlim([2,5])
Axes[1].grid(True)
```

# Other 2D Graphics

```python
n=np.array([1,2,3,4,5,6])
fig,axes=plt.subplots(1,4,figsize=(16,5))

axes[0].set_title('scatter')
axes[0].scatter(x,x+0.25*np.random.randn(len(x)))

axes[1].set_title('step')
axes[1].step(n,n**2,lw=2)

axes[2].set_title('bar')
axes[2].bar(n,n**2,align='center',width=0.5,alpha=0.5)

axes[3].set_title('fill_between')
axes[3].fill_between(x,x**2,x**3,color='green',alpha=0.5)
```

```python
'Draw a radar chart'
fig=plt.figure(figsize=(6,6))
ax=fig.add_axes([0.0,0.0,.6,.6],polar=True) # Polar change de type of graph from rectangule to circular
t=np.linspace(0,2*np.pi,100)
ax.plot(t,t*2,color='blue',lw=3)
```

```python
'Draw histogram'
n=np.random.randn(100000)
fig,axes=plt.subplots(1,2,figsize=(12,4))

axes[0].set_title('Default histogram')
axes[0].hist(n)

axes[1].set_title('Cumulative detailed histogram')
axes[1].hist(n,cumulative=True,bins=10)
```

```python
# Draw contour image

import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta=0.025
x=np.arange(-3.0,3.0,delta)
y=np.arange(-2.0,2.0,delta)
X,Y=np.meshgrid(x,y)
Z1=np.exp(-X**2-Y**2)
Z2=np.exp(-(X-1)**2-(Y-1)**2)
Z=(Z1-Z2)*2

fig,ax=plt.subplots()
CS=ax.contour(X,Y,Z)
ax.clabel( ,inline=1,fontsize=10)
ax.set_title('contour map')
```

```python
# Draw 3D surface image
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig=plt.figure(figsize=(14,6))

#Specify the 3D graphics to draw, with projection ='3d'
ax=fig.add_subplot(1,2,1,projection='3d')
ax.plot_surface(X,Y,Z,rstride=4,cstride=4,linewidth=0)
```

# Write a Python programming to create a pie chart of the popularity of programming languages

```python
import matplotlib.pyplot as plt
# Data to plot
languages='Python','Java','PHP','JavaScript','C#','C++'
popularity=[31.8,24.6,12.3,11.2,10.8,9.4]
colors=['#124578','#986532','#13ab46','#79cd46','#13ef46','#46aa79']

# Explode 1st slice
explode=(0.2,0,0,0,0,0)

fig,ax=plt.subplots()
ax.pie(popularity,explode=explode,labels=languages,colors=colors,autopct='%1.1f%%',shadow=True,startangle=0)
ax.axis('equal') 
plt.show()
```
